```python
import pandas as pd
import numpy as np 
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go
from ipywidgets import widgets, interactive
# import plotly.plotly as py

GDP_Life = pd.read_csv('GDP vs Life.csv',header=0)
display(GDP_Life)
```

```python
GDP_Life.drop(GDP_Life.index[[434, 435,436,437,438]],inplace=True)
GDP_Life= GDP_Life.replace('..',0)
display(GDP_Life)
```

```python
GDP=[]
Life=[]

GDP =GDP_Life[GDP_Life['Series_Name']=="GDP per capita (current US$)"]
GDP=GDP.set_index('Country_Name',drop=True)
GDP.drop(GDP.columns[[0,1,2]], axis=1 ,inplace=True)
display(GDP)

```

```python
Life = GDP_Life[GDP_Life['Series_Name']=="Life expectancy at birth, total (years)"]
Life=Life.set_index('Country_Name',drop=True)
Life.drop(Life.columns[[0,1,2]], axis=1 ,inplace=True)
display (Life)
```

```python
Country=GDP_Life['Country_Name'].unique()
#print('Country',Country)
#print('Test',GDP.loc[Country[5]])
Years=list(range(1960,2017))

CountryName = widgets.Dropdown(
    options=Country,
    value='Angola',
    description='Country:',
)


def plot_scatterplot(CountryName):
    Y1_data=GDP.loc[CountryName]
    Y2_data=Life.loc[CountryName]
    trace1 = go.Scatter(
        x=Years,
        y=Y1_data,
        mode='lines+markers',
        name='GDP per capita (current US$)',
        marker=dict(
            color='rgba(48, 112, 99, 1)',
            line=dict(
                color='rgba(48, 112, 99, 1)',
                width=1,
            ),
            symbol='circle',
            size=8,
        )
    )
    trace2 = go.Scatter(
       x=Years,
        y=Y2_data,
        mode='lines+markers',
        name='Life expectancy at birth, total (years)',
        yaxis='y2',
        marker=dict(
            color='rgba(153, 0, 0, 1)',
            line=dict(
                color='rgba(153, 0, 0, 1)',
                width=1,
            ),
            symbol=104,
            size=8,
        )
    )


    data = [trace1, trace2]

    layout = go.Layout(
        title='<b>GDP per capita (current US$) VS Life expectancy at birth, total (years)</b>',
        yaxis=dict(
            title='GDP per capita (current US$)',
            titlefont=dict(
                color=' rgb(48, 112, 99)',
                size=16,
            ),
            tickfont=dict(
                color=' rgb(48, 112, 99)'
             ),
        ),
        yaxis2=dict(
            title='Life expectancy at birth, total (years)',
            titlefont=dict(
                color='rgba(153, 0, 0, 1)',
                size=16,
            ),
            tickfont=dict(
                color='rgba(153, 0, 0, 1)'
            ),
            overlaying='y',
            side='right'
            ),
        legend=dict(x=1,y=1.13 ,bgcolor='#E2E2E2',bordercolor='#FFFFFF',borderwidth=2)

    )

    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig, validate=False, filename='GDP_vs_Life.html')
    
    
interactive(plot_scatterplot,CountryName=CountryName)
```
