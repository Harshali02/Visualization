```python
import pandas as pd
import numpy as np 
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go
GDP_Current = pd.read_csv('GDP_Current.csv',header=0)
GDP_PPP = pd.read_csv('GDP_PPP.csv',header=0)
GDP_Per_Capita = pd.read_csv('GDP_Per_Capita_current.csv',header=0)
display(GDP_Current)
```

```python
new_names = {'2016 [YR2016]':'2016',
            'Country Name':'Country'}
GDP_Current.rename(columns = new_names,inplace=True)
GDP_Current.drop(GDP_Current.index[[217,218,219,220,221]],inplace=True)
GDP_Current= GDP_Current.replace('..',0)
display(GDP_Current)
```

```python
new_names = {'2016 [YR2016]':'2016',
            'Country Name':'Country'}
GDP_PPP.rename(columns = new_names,inplace=True)
GDP_PPP.drop(GDP_PPP.index[[217,218,219,220,221]],inplace=True)
GDP_PPP= GDP_PPP.replace('..',0)
display(GDP_PPP)
```

```python
display(GDP_Per_Capita)
```

```python
new_names = {'2016 [YR2016]':'2016',
            'Country Name':'Country'}
GDP_Per_Capita.rename(columns = new_names,inplace=True)
GDP_Per_Capita.drop(GDP_Per_Capita.index[[217,218,219,220,221]],inplace=True)
GDP_Per_Capita= GDP_Per_Capita.replace('..',0)
display(GDP_Per_Capita)
```

```python
GDP__Current = go.Choropleth(
        locationmode = 'country names',
        locations = GDP_Current['Country'],
        z = GDP_Current['2016'],
#         text = df['Name'] ,
        colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],\
            [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
       # hoverinfo='text+z',
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
#             autotick = False,
            thickness=15,
            title = 'GDP_Current'),
        geo = 'geo1'
         )


GDP__PPP = go.Choropleth(
        locationmode = 'country names',
        locations = GDP_PPP['Country'],
        z = GDP_PPP['2016'],
       # text = GDP_PPP['Country'] ,
        colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],\
            [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        #hoverinfo='text+z',
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
         colorbar = dict(
          thickness =15,
        title = 'GDP_PPP'),
        geo = 'geo1'
         )

GDP__PER__Capita = go.Choropleth(
        locationmode = 'country names',
        locations = GDP_Per_Capita['Country'],
        z = GDP_Per_Capita['2016'],
       # text = GDP_PPP['Country'] ,
        colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],\
            [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        #hoverinfo='text+z',
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
         colorbar = dict(
          thickness =15,
          title = 'GDP_Per_Capita'),
        geo = 'geo1'
         )

data = [GDP__Current, GDP__PPP, GDP__PER__Capita]
updatemenus = list([
    dict(
         buttons=list([
             dict(
                 label='GDP_Current',
                  method='update',
                  args=[{'visible': [True, False, False]}, ]
                  ),
             dict(label='GDP_PPP',
                  method='update',
                  args=[{'visible': [False, True, False]}, ]
                  ),
              dict(
                 label='GDP_Per_Capita',
                  method='update',
                  args=[{'visible': [False, False, True]}, ]
                  ),
         ]),
         direction = 'down',
         showactive = True,
         x = 0.3,
         xanchor = 'left',
         y = 0.1,
         yanchor = 'bottom',
        # bgcolor = 'white',
         bordercolor = '#41B3A3',
         font = dict(size=11)
    ),
])

layout = go.Layout(
        title = '<b> Gross Domestic Product, 2016</b>', 
        titlefont= dict(size= 24,
                        color='rgb(0,0,0)',),
        
        geo1 = dict(
            resolution = 110,
            showframe = False,
            showcoastlines = False,
            projection = dict(
                type = 'mercator'
            ),
        ),
        
        legend = dict(
               traceorder = 'reversed',
                x=0,
                y=0,

        ),
    
)
annotations = list([
    dict(text='<b>Select Type</b>', x=0.35, y = 0.18, yref='paper', align='left', showarrow=False,font=dict(size=11,)),
    dict(text='<b>A world map that indicates the Gross Domestic Product for year 2016</b>', x=0.7, y = 1.075, yref='paper', align='center', showarrow=False,font=dict(size=11,)),
])

layout['updatemenus'] = updatemenus
layout['annotations'] = annotations

fig = go.Figure(data=data, layout=layout)
py.plot(fig, validate=False, filename='world_map1.html')
```
